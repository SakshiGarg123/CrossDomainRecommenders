```python
Rev=(data.Reviews)
Rev=Rev.dropna().reset_index()
Rev=(Rev.Reviews).tolist()
Story=(data.Storyline).tolist()
type(Rev)
```

```python
from gensim import corpora
stoplist = set('for a of the and to up in i have been an Imdb never NaN it is was \n \\n before after that am this my he today what are be would can not no ever can\'t all so'.split())
texts = [[word for word in text.lower().split() if word not in stoplist]for text in Rev]       
texts[1]
```

```python
from collections import defaultdict
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1

texts = [[token for token in text if (frequency[token] > 1 and frequency[token] < 4)] for text in texts]

from pprint import pprint  # pretty-printer
pprint(len(texts[0]))
```

```python
dictionary = corpora.Dictionary(texts)
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
for c in corpus:
    print(c)
```
